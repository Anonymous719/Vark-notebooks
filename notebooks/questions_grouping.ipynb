```python
def drop_columns_by_index(df, columns_to_drop):
    """
    Drop specified columns from a dataframe based on their index positions.

    Parameters:
    df (pandas.DataFrame): The dataframe from which to drop columns.
    columns_to_drop (list): A list of column index positions to drop.

    Returns:
    pandas.DataFrame: The dataframe with the specified columns dropped.
    """
    columns_to_keep = [col for col in df.columns if df.columns.get_loc(col) not in columns_to_drop]
    df_dropped = df[columns_to_keep]
    return df_dropped
```

```python
df = drop_columns_by_index(df, [0, 1, 2, 3, 4, 5, 7, 8])
df
```

```python
def truncate_column_titles(df, start_col=1):
    """
    Truncate the titles of columns in a dataframe to the first 6 characters, starting from a specified column.

    Parameters:
    df (pandas.DataFrame): The dataframe to modify.
    start_col (int): The index of the first column to modify. Default is 4.

    Returns:
    pandas.DataFrame: The dataframe with truncated column titles.
    """
    df_truncated = df.copy()
    df_truncated.columns = [col[:3] if i >= start_col else col for i, col in enumerate(df_truncated.columns)]
    return df_truncated
```

```python
df = truncate_column_titles(df)
df
```

```python
positive_df['Average'] = positive_df.iloc[:, 1:].mean(axis=1, skipna=True)
negative_df['Average'] = negative_df.iloc[:, 1:].mean(axis=1, skipna=True)
neutral_df['Average'] = neutral_df.iloc[:, 1:].mean(axis=1, skipna=True)
```

```python
positive_floor = positive_df['Average'].apply(lambda x: int(x))
negative_floor = negative_df['Average'].apply(lambda x: int(x))
neutral_floor = neutral_df['Average'].apply(lambda x: int(x))

# Create histogram plot
plt.figure(figsize=(10, 6))
plt.hist([positive_floor, negative_floor, neutral_floor], bins=range(1, max(max(positive_floor), max(negative_floor), max(neutral_floor)) + 1), alpha=0.7, label=['Positive', 'Negative', 'Neutral'], edgecolor='black')
plt.xlabel('Floor Value of Average')
plt.ylabel('Frequency')
plt.title('Histogram of Floor Value of Average')
plt.legend()
plt.grid(True)
plt.show()
```

```python
result_df = pd.DataFrame(positive_df['ID'])

# Fill 'Positive_Mean' column
result_df['Positive_Mean'] = positive_df['Average']

# Fill 'Negative_Mean' column
result_df['Negative_Mean'] = negative_df['Average']

# Fill 'Neutral_Mean' column
result_df['Neutral_Mean'] = neutral_df['Average']

# Display the resulting DataFrame
print(result_df)
```

```python
column = "Neutral_Mean"
# Combine positive, negative, and neutral averages into a single overall average
result_df['Overall_Average'] = result_df[['Positive_Mean', 'Negative_Mean', 'Neutral_Mean']].mean(axis=1)

# Calculate percentiles based on the overall average
percentiles = np.percentile(result_df[column], [20, 40, 60, 80])

# Define category names
categories = ['Poor', 'Below Average', 'Average', 'Above Average', 'Excellent']

# Divide students into categories based on percentiles of overall average
result_df['Category'] = pd.cut(result_df[column], bins=[-np.inf] + list(percentiles) + [np.inf], labels=categories, include_lowest=True)

# Count the frequency of each category
category_counts = result_df['Category'].value_counts()

print("Overall Average Categories:")
print(category_counts)
```

```python
students_df = pd.read_csv("./../survey/familarity_with_online/GR.csv")
```

```python
students_df = students_df.loc[:, ['ID', 'Full name']]
students_df
```

```python
students_df['Name'] = students_df['Full name'].str.lower()
grade_df['Name'] = grade_df['Name'].str.lower()

# Merge the DataFrames based on the 'Name' column
merged_df = pd.merge(students_df, grade_df, on='Name', how='inner')

# Drop the extra 'Name' column created during merge
merged_df.drop(columns=['Name'], inplace=True)

# Display the merged DataFrame
print(merged_df)
```

```python
# Drop all columns except 'ID' and 'Class'
merged_df = merged_df[['ID', 'Class']]

# Display the updated DataFrame
print(merged_df)
```

```python
# Merge the DataFrames based on the 'ID' column
combined_df = pd.merge(result_df, merged_df, on='ID', how='inner')

# Display the combined DataFrame
print(combined_df)
```

```python
# Calculate the number of matching values between 'Class' and 'Category'
matching_count = (combined_df['Class'] == combined_df['Category']).sum()

# Calculate the total number of rows in the DataFrame
total_count = combined_df.shape[0]

# Calculate the accuracy percentage
accuracy_percentage = (matching_count / total_count) * 100

# Print the accuracy percentage
print(f"Accuracy Percentage: {accuracy_percentage:.2f}%")
```

```python
# Count the frequency of each category
category_counts = combined_df['Class'].value_counts()

# Plot the histogram
plt.figure(figsize=(8, 6))
plt.bar(category_counts.index, category_counts.values, color='skyblue')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.title('Histogram of Student Categories')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
