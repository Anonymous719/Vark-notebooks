```python
for col in df.columns[2:]:
    df[col] = df[col].str[0]

df
```

```python
# Define the VARK matrix
vark_matrix = np.array([
    [4, 2, 3, 1],
    [1, 2, 3, 4],
    [2, 1, 4, 3],
    [3, 2, 4, 1],
    [2, 1, 4, 3],
    [3, 4, 2, 1],
    [3, 2, 4, 1],
    [4, 3, 1, 2],
    [4, 2, 1, 3],
    [2, 4, 3, 1],
    [1, 3, 2, 4],
    [3, 4, 2, 1],
    [4, 2, 3, 1],
    [4, 3, 2, 1],
    [4, 2, 3, 1],
    [1, 2, 3, 4]
])

# Define a function to calculate the VARK score for each student
def calculate_vark_score(row):
    visual = 0
    aural = 0
    read_write = 0
    kinesthetic = 0
    
    for i in range(1, 17):
        if row[f'Q{i}'] == '1':
            visual += vark_matrix[i-1][3]
        elif row[f'Q{i}'] == '2':
            aural += vark_matrix[i-1][1]
        elif row[f'Q{i}'] == '3':
            read_write += vark_matrix[i-1][2]
        elif row[f'Q{i}'] == '4':
            kinesthetic += vark_matrix[i-1][0]
    
    return pd.Series([visual, aural, read_write, kinesthetic])

# Apply the function to the DataFrame
vark_scores = df.apply(calculate_vark_score, axis=1)

# Assign the student to the VARK group with the highest score
df['vark_div'] = vark_scores.idxmax(axis=1)
```
